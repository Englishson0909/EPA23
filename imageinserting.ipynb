```python
#@markdown Install and import libraries {gtts},{Audio}
!pip install gtts
from gtts import gTTS
from IPython.display import Audio, display

def tts(txt):
  tts = gTTS(txt, lang='en', tld="us")
  tts.save("sample.mp3")
  return Audio("/content/sample.mp3")
```

```python
#@markdown Test: audio

text = "I dropped the bowl."
tts(text)
Audio("sample.mp3", autoplay=True)
```

```python
#@markdown Listen to the following sentence and generate an image illustrating the situation.
import requests
import io

url = 'https://raw.github.com/Englishson0909/EPA23/main/Once%20upon%20a%20time.txt'
response = requests.get(url)

text_data = io.StringIO(response.content.decode('utf-8'))

# Print the contents of the file
mytext = text_data.read()

mytext
tts(mytext)
Audio("sample.mp3")
```
