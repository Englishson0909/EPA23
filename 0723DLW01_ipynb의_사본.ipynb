```python
#@markdown üîÑ Making functions: etts('text'), ktts('text')
%%capture
!pip install gTTS
from gtts import gTTS
from IPython.display import Audio

dialect = "us"
def etts(text):
  text_to_say = text

  gtts_object = gTTS(text = text_to_say,
                     lang = "en", tld = dialect,
                    slow = True)

  gtts_object.save("E-audio.mp3")
  return Audio("E-audio.mp3")

def ktts(text):
  text_to_say = text

  gtts_object = gTTS(text = text_to_say,
                     lang = "ko",
                    slow = False)

  gtts_object.save("K-audio.mp3")
  return Audio("K-audio.mp3")
```

```python
points = "t3" #@param ["t1", "t2", "t3", "t4", "t5","t6"]

texts = "Clear Communication Good pronunciation helps you speak clearly and be easily understood by others. Better Listening Learning pronunciation improves your ability to understand what others are saying. Confidence Boost Knowing how to pronounce words correctly makes you feel more confident when speaking. Cultural Appreciation Pronunciation connects you to the culture of the language you are learning. Professional Benefits Proper pronunciation is valuable in job interviews and professional settings. Friendlier Conversations When you pronounce words clearly, it creates a friendly and welcoming atmosphere during conversations, making it easier for others to engage with you."



sents = texts.split(".")

if points == "t1":
  txt = sents[0]
elif points == "t2":
  txt = sents[1]
elif points == "t3":
  txt = sents[2]
elif points == "t4":
  txt = sents[3]
elif points == "t5":
  txt = sents[4]
elif points == "t6":
  txt = sents[5]

etts(txt)
Audio("E-audio.mp3", autoplay = True)
```
